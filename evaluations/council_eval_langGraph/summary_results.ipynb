```python
import json

def analyze_recommendation_systems(data: list):
    """分析兩個推薦系統的表現"""
    buildmate_wins = 0
    gpt4o_wins = 0
    buildmate_total = 0
    gpt4o_total = 0
    valid_questions = len(data)
    
    for q in data:
        
        print (q['winner'])
        if q.get('winner') == 1:
            buildmate_wins += 1
        else:
            gpt4o_wins += 1
        
        buildmate_total += q.get('buildmate_avg_score', 0)
        gpt4o_total += q.get('gpt4o_avg_score', 0)
    
    # 計算平均分
    buildmate_avg = buildmate_total / valid_questions if valid_questions > 0 else 0
    gpt4o_avg = gpt4o_total / valid_questions if valid_questions > 0 else 0
    
    print(f"\n分析了 {valid_questions} 個問題")
    
    print("\nNumber of wins:")
    print(f"BuildMate: {buildmate_wins}")
    print(f"GPT4o: {gpt4o_wins}")
    
    print("\nAverage score:")
    print(f"BuildMate: {buildmate_avg:.2f}")
    print(f"GPT4o: {gpt4o_avg:.2f}")
    
    print("\nTotal score:")
    print(f"BuildMate: {buildmate_total:.2f}")
    print(f"GPT4o: {gpt4o_total:.2f}")

def main():
    try:
        with open('evaluation_report_budget_reason_gemini_all_app_2grader.json', 'r', encoding='utf-8') as f:
            data = json.load(f)
        analyze_recommendation_systems(data)
    except FileNotFoundError:
        print("錯誤：找不到數據文件")
    except json.JSONDecodeError:
        print("錯誤：JSON 文件格式不正確")
    except Exception as e:
        print(f"發生錯誤：{str(e)}")

if __name__ == "__main__":
    main()
```
